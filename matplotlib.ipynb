```python

age = np.arange(25, 44, 2.5)
salary = np.arange(40000, 76000, 5000)
salary_2 = [45372, 46852, 69401, 56790, 73000, 75000, 80000, 83454]
```

```python
plt.figure(figsize=(10,5))

plt.plot(age, salary, "b:", label = "SALARY")
plt.plot(age, salary_2, "r-", label = "SALARY 2")

plt.xlabel("AGE")
plt.ylabel("SALARY")
plt.title("SALARY BY AGE")

plt.legend()
plt.show()


#plt legend demezsek grafik çizer ama hangi çizgi hangi salary nin
#onu göstermez.
```

```python
plt.figure(figsize=(10,5))

plt.plot(age, salary, "b:", label = "SALARY")
plt.plot(age, salary_2, "r-", label = "SALARY 2")

plt.xlabel("AGE")
plt.ylabel("SALARY")
plt.title("SALARY BY AGE")

plt.legend(loc = "lower right")
plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.plot(age, salary, "b:", label = "SALARY")
plt.plot(age, salary_2, "r-", label = "SALARY 2")

plt.xlabel("AGE")
plt.ylabel("SALARY")
plt.title("SALARY BY AGE")

plt.legend(loc = 5)
plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.plot(age, salary, "b:", label = "SALARY")
plt.plot(age, salary_2, "r-", label = "SALARY 2")

plt.xlabel("AGE")
plt.ylabel("SALARY")
plt.title("SALARY BY AGE")

plt.legend(bbox_to_anchor = (0, 0, 1, 1))
plt.show()
```

```python
plt.figure(facecolor="yellow", figsize = (10, 5))
ax = plt.axes()
ax.set_facecolor("xkcd:mint green")
plt.plot(age, salary, "b-", linewidth = 4, label = "SALARY 1")
plt.plot(age, salary_2, "r", linestyle  = ":", label = "SALARY 2")
plt.xlabel("AGE")
plt.ylabel("SALARY")
plt.title(label = "SALARY BY AGE")
plt.legend(bbox_to_anchor = (0.5, 0.5, 0.5, 0.5))
plt.show()
```
